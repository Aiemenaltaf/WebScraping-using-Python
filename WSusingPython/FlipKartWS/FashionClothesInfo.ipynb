```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

#install chrom webdriver
driver = webdriver.Chrome(executable_path=r'C:\Users\Altaf Hussain\Downloads\chromedriver_win32\chromedriver.exe')

# lists to store attributes of products
products=[] 
prices=[] 
stars =[]
sizes=[]
brands=[]
product_detail=[]
colors=[]
sales=[]

# loop to store data in csv file
for page in range(1,26):
    driver.get("https://www.flipkart.com/clothing-and-accessories/pr?sid=clo&marketplace=FLIPKART&p%5B%5D=facets.ideal_for%255B%255D%3DWomen&otracker=nmenu_sub_Women_0_Clothing&page"+str(page))
    content = driver.page_source
    soup = BeautifulSoup(content)
    for a in soup.findAll('div',attrs={'class':'_1xHGtK _373qXS'}):
        name=a.find('a', attrs={'class':'IRpwTa'})
        products.append(name.text)
        price=a.find('div', attrs={'class':'_25b18c'})
        prices.append(price.text)
        if(a.find('div', attrs={'class':'_2WkVRV'})):
            brand=a.find('div', attrs={'class':'_2WkVRV'})
            brands.append(brand.text)
        else:
            brand="No brand Available"
            brands.append(brand)
        if(a.find('div', attrs={'class':'_3eWWd-'})):
                color=a.find('div', attrs={'class':'_3eWWd-'})
                colors.append(color.text)
        else:
            color="No color Available"
            colors.append(color)
        if(a.find('div', attrs={'class':'_1AtVbE col-12-12'})):
            prod_detail=a.find('div', attrs={'class':'_1AtVbE col-12-12'})          
            product_detail.append(prod_detail.text)
        else:
            prod_detail="No detail Available"
            product_detail.append(prod_detail)
        if(a.find('div', attrs={'class':'_22QfJJ'})):
            size=a.find('div', attrs={'class':'_22QfJJ'}).text
            sizes.append(size)
        else:
            size="No size Available"
            sizes.append(size)
        if(a.find('div', attrs={'class':'_3Ay6Sb'})):
            sale=a.find('div', attrs={'class':'_3Ay6Sb'}).text
            sales.append(sale)
        else:
            sale="No sale Available"          
            sales.append(sale) 
 
       # print length of data
    print(len(products))
    print(len(prices))
    print(len(sizes))
    print(len(brands))
    print(len(product_detail))
    print(len(colors))
    print(len(sales))
        
    df = pd.DataFrame({'Product Name':products,'Price':prices,'Size':sizes, 'Brand':brands, 'Description':product_detail, 'Color': colors,'Sale':sales})
    df.to_csv('Fashion.csv', index=False, encoding='utf-8')
    dataset = pd.DataFrame(data = df)
    print(dataset)

```
